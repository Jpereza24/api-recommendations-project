```python
def user_text(lista):
    users_dict=dict()
    for i in range(len(lista)):
        if lista[i]['userName'] not in users_dict:
            users_dict[lista[i]['userName']]=lista[i]['text']
        else:
            users_dict[lista[i]['userName']]+=' ' +lista[i]['text']
    for e in users_dict:
        users_dict[e]=re.sub(r"[^a-zA-Z0-9]+", ' ', users_dict[e])
    return users_dict
```
